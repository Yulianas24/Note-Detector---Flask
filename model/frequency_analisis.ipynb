```python
import librosa 
import librosa.display 
import os
import matplotlib.pyplot as plt
import IPython.display as ipd
import pandas as pd
import csv
import numpy as np
```

```python
# mengambil file audio
low = "Notes Datasets/3. F-sharp/03-F-sharp (1).wav"
high = "Notes Datasets/4. G/04-G-7.wav"
```

```python
signal_low, sr = librosa.load(low)
signal_high, sr = librosa.load(high)
```

```python
ipd.Audio(data = signal_low, rate = sr)
```

## Menampilkan grafik time domain

```python
def show_signal(signal):
    plt.figure(figsize=(25,10))
    plt.title("data")
    plt.plot(signal)
    plt.show()

show_signal(signal_low)
```

## Trimming Audio

```python
signal_low, _ = librosa.effects.trim(signal_low, top_db=1)
signal_high, _ = librosa.effects.trim(signal_high, top_db=1)
show_signal(signal_low)
```

```python
ipd.Audio(data = signal_low, rate = sr)
```

## Hitung short time furier transform 

```python
fft_low = librosa.stft(signal_low, center = False)
fft_high = librosa.stft(signal_high, center = False)
energi_low = abs(fft_low)**2
energi_high = abs(fft_high)**2
```

```python
def transform(energi):
  data = energi[:70]
  plt.figure(figsize=(25,10))
  plt.title("stft")
  plt.plot(data)
  plt.show()
transform(energi_low)
```

```python
transform(energi_high)
```
