```python
import librosa
import numpy as np
import os
import pandas as pd
PATH = 'Notes Datasets/'
datas = []


```

```python
def pitch(audio, label):
  y, sr = librosa.load(audio)
  signal, _ = librosa.effects.trim(y, top_db=2)
  fft = librosa.stft(signal, center = False)
  energi = abs(fft)**2
  data = pd.DataFrame(energi[:70])
  index = data.max().idxmax()
  formatted_data = map(lambda x: format(x, '.2f'), data[index])
  data =  list(formatted_data)
  data.append(label)
  return data


```

```python
print(pitch('Notes Datasets/04. G/04-G-2.wav', 'E'))
```

```python
for i in os.listdir(PATH):
  for j in os.listdir(f'{PATH}/{i}'):
    # print(f'{j}')
    datas.append(pitch((f'{PATH}/{i}/{j}'), i[4:]))
```

```python
df = pd.DataFrame(datas)
df 
```

```python
df.to_csv('data.csv', index=False)
```
